```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns 
import os
```

```python
#Loading and reading csv data from local files
csv = pd.read_csv("D:\Work Datasets\Cleaned Datasets\cleaned_data_2.csv")
```

```python
target_data['dob'] = pd.to_datetime(target_data['dob'], errors='coerce')
```

```python
target_data.dtypes
```

```python
target_data['dob'] = target_data['dob'].apply(lambda x: x.strftime('%Y-%m-%d')if not pd.isnull(x) else '')
```

```python
target_data['dob'].astype('datetime64')
```

```python
import datetime as dt

cur_date = dt.datetime.now()
target_data['Age'] = target_data['dob'].apply(lambda x: len(pd.date_range(start = x, end = cur_date, freq = 'Y')) if str(x) else None)
```

```python
#Did not work
target_data['Age'] = target_data['Age'].fillna(target_data['Age'].mode())
```

```python
# replacing NaN values with 0(did not work)
target_data['Age'] = target_data['Age'].fillna(0)
 
target_data['Age'].astype('int64')
```

```python
target_data['Age'].mode()
```

```python
#Replacing the values with 0 with the mode of Age column
mode = target_data['Age'].mode()
target_data['Age'] = target_data['Age'].replace(to_replace = [0],value = mode)
#target_data['Age'] = target_data['Age'].fillna(target_data['Age'].mode())
```

```python
target_data.isnull().sum()
```

```python
#now there's one condition for the Age column
target_data = target_data[target_data.Age < 100]
```

# Now some insights about the New Customers....

```python
plt.figure(figsize = (12,8))
plt.hist(target_data['Age'], color = 'deepskyblue', edgecolor = 'white', bins=[0,24,44,64,100])
plt.title('Age distribution histogram')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.countplot(x = 'gender', data = target_data)
plt.title('Gender count barchart')
plt.show()
```

```python
plt.figure(figsize = (12,8))
sns.catplot(x='gender', data=target_data, y='Age')
#plt.legend(title='Gender', loc='upper right', fancybox=True)
plt.title('Age Distribution Across Gender')
plt.show()
```

```python
plt.figure(figsize = (12,8))
sns.countplot(x='gender', data=target_data, hue='Age')
plt.legend(title='Gender', loc='upper right', fancybox=True)
plt.title('Age Distribution Across Gender')
plt.show()
```

```python
plt.figure(figsize = (16,8))
sns.countplot(x='job_industry_category', data=target_data, hue='gender')
plt.legend(title='Gender', loc='upper right', fancybox=True)
plt.title('Job industry category across gender')
plt.show()
```

```python
plt.figure(figsize = (12,8))
sns.boxplot(x = target_data['gender'], y = target_data['Age'])
plt.title('Boxplot showing the age range of customers')
plt.show()
```

```python
plt.figure(figsize = (16,8))
sns.countplot(x='state', data=target_data, hue='property_valuation')
plt.legend(title='State', loc='upper right', fancybox=True)
plt.title('Property Valuation across state')
plt.show()
```

```python
plt.figure(figsize = (16,8))
sns.countplot(x='wealth_segment', data=target_data, hue='property_valuation')
plt.legend(title='State', loc='upper right', fancybox=True)
plt.title('Property Valuation across Wealth Segment')
plt.show()
```
